```python
def load_metrics(acc_form, roc_auc_form, acc_dir_path, roc_auc_dir_path, sizes, folds=5):
    data = {}
    for size in sizes:
        data[str(size)] = {}
        for fold in range(folds):
            print(fold)
            json_acc_path = os.path.join(acc_dir_path, acc_form.format(size, fold))
            json_roc_auc_path = os.path.join(roc_auc_dir_path, roc_auc_form.format(size, fold))
            data[str(size)][str(fold)] = {'accuracy': get_json_data(json_acc_path)['accuracy'], 'auc': get_json_data(json_roc_auc_path)['auc']}
    return data

def load_metrics1(acc_form, roc_auc_form, acc_dir_path, roc_auc_dir_path, sizes):
    data = {}
    for size in sizes:
        data[str(size)] = {}
        json_acc_path = os.path.join(acc_dir_path, acc_form.format(size))
        json_roc_auc_path = os.path.join(roc_auc_dir_path, roc_auc_form.format(size))
        data[str(size)] = {'accuracy': get_json_data(json_acc_path)['accuracy'], 'auc': get_json_data(json_roc_auc_path)['auc']}
    return data
```

```python
def get_plot_df(model_metrics):
    count = 0
    df = pd.DataFrame(columns=['value', 'metric', 'fold'])
    for str_fold, metrics_dict in model_metrics.items():
        fold = int(str_fold) + 1 
        for metric in ['accuracy', 'auc']:
            df.loc[count] = [model_metrics[str_fold][metric], metric, fold]
            count += 1
    return df

def get_plot_df2(model_metrics):
    count = 0
    df = pd.DataFrame(columns=['value', 'metric', 'fold'])
    for str_fold, metrics_dict in model_metrics.keys():
        ep = int(str_fold) + 1 
        for metric in ['accuracy', 'auc']:
            df.loc[count] = [model_metrics[str_fold][metric], metric, fold]
            count += 1
    return df

def plot_fold_bar(data):
    plot_df = get_plot_df(data)
    plt.figure(figsize=(10, 7))
    ax = sns.barplot(x="fold", y="value", hue="metric", data=plot_df)
    plt.grid('on')
```

```python
def print_num_fn(df1, df2):
    conf_1 = df1['max_conf']
    conf_2 = df2['max_conf']
    print(f'1) fn num - {len(df1[conf_1 == 0])}') 
    print(f'2) fn num - {len(df2[conf_2 == 0])}') 

def print_low_iou(df1, df2):
    len_1_iou = len(df1[df1['iou'] < 0.75])
    len_2_iou = len(df2[df2['iou'] < 0.75])
    print(f'1) num low_iou - {len_1_iou}') 
    print(f'2) num low_iou - {len_2_iou}')
```
