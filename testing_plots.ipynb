```python
# Пути к json-файлам
m2500_json_path = 'testing_files/data/result/json_roc_auc_data/roc_auc_data_model_32_bg2500.json'
m5000_json_path = 'testing_files/data/result/json_roc_auc_data/roc_auc_data_model_32_bg5000.json'
m8000_json_path = 'testing_files/data/result/json_roc_auc_data/roc_auc_data_model_32_bg5000.json'
m10000_json_path = 'testing_files/data/result/json_roc_auc_data/roc_auc_data_model_32_bg10000.json'
m12000_json_path = 'testing_files/data/result/json_roc_auc_data/roc_auc_data_model_32_bg12000.json'

# данные с json-ов
data_m2500 = get_json_data(m2500_json_path)
data_m5000 = get_json_data(m5000_json_path)
data_m8000 = get_json_data(m8000_json_path)
data_m10000 = get_json_data(m10000_json_path)
data_m12000 = get_json_data(m12000_json_path)

# Значения AUC для каждой модели
auc_m2500 = data_m2500['auc']
auc_m5000 = data_m5000['auc']
auc_m8000 = data_m8000['auc']
auc_m10000 = data_m10000['auc']
auc_m12000 = data_m12000['auc']
```

```python
plt.figure(figsize=(15, 10))
plt.plot(data_m2500['fpr'], data_m2500['tpr'], label=f'm2500 (roc: {auc_m2500})');
plt.plot(data_m5000['fpr'], data_m5000['tpr'], label=f'm5000 (roc: {auc_m5000})');
plt.plot(data_m8000['fpr'], data_m8000['tpr'], label=f'm8000 (roc: {auc_m8000})');
plt.plot(data_m10000['fpr'], data_m10000['tpr'], label=f'10000 (roc: {auc_m10000})');
plt.plot(data_m12000['fpr'], data_m12000['tpr'], label=f'12000 (roc: {auc_m12000})');
plt.legend()
plt.title('ROC-AUC')
plt.grid('on')
plt.savefig('testing_files/data/result/roc_auc.jpg')
plt.show();
```
